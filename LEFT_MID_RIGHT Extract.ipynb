```python
import pandas as pd
import numpy as np
```

```python
# Importing my data

df = pd.read_csv("car inventory.txt")
df
```

```python
# The lower part of the dataset is empty, so we drop where all columns are null

df.dropna(axis = 0, how = 'all', inplace= True)
df
```

```python
# 1. The first 2 letters is the Make, extract it to the Make column e.g FD
df.Make = df['Car ID'].str[:2]
df
```

```python
# 2. From the Make column, generate the Make Full Name, using this table below e.g Ford

# To do this, there are many methods, I used numpy select function


condition = [df.Make == 'CR', df.Make == 'FD', df.Make == 'GM', df.Make == 'HO', 
             df.Make == 'HY', df.Make == 'TY']

choice = ['Chrysler', 'Ford', 'General Motors', 'Honda', 'Hyundai','Toyota']


df['Make (Full Name)'] = np.select(condition, choice)
df
```

```python
# From the Car ID column, generate the Model column, the 5th to 7th letters represents the Model of the car e.g MTG

df.Model = df['Car ID'].str[4:7]
df
```

```python
# From the Model column, generate the Model Full Name using the table below; e.g Mustang

# Here I used another method; the For loop. 

for i in df['Model']:
    if i == 'CAM':
        df['Model (Full Name)'] = 'Camry'
    elif i == 'CAR':
        df['Model (Full Name)'] = 'Caravan'
    elif i == 'CIV':
        df['Model (Full Name)'] = 'Civic'
    elif i == 'CMR':
        df['Model (Full Name)'] = 'Camero'
    elif i == 'COR':
        df['Model (Full Name)'] = 'Corolla'
    elif i == 'ELA':
        df['Model (Full Name)'] = 'Elantra'
    elif i == 'FCS':
        df['Model (Full Name)'] = 'Focus'
    elif i == 'MTG':
        df['Model (Full Name)'] = 'Mustang'
    elif i == 'ODY':
        df['Model (Full Name)'] = 'Odyssey'
    elif i == 'PTC':
        df['Model (Full Name)'] = 'PT Cruiser'
    elif i == 'SLV':
        df['Model (Full Name)'] = 'Silvarado'
        
        
df


# But I didn't get the desired result for every datapoint was 'Elantra'
```

```python
# Another method to make our for loop work.......


model = []

for i in df['Model']:
    if i == 'CAM':
        model.append('Camry')
    elif i == 'CAR':
        model.append('Caravan')
    elif i == 'CIV':
        model.append('Civic')
    elif i == 'CMR':
        model.append('Camero')
    elif i == 'COR':
        model.append('Corolla')
    elif i == 'ELA':
        model.append('Elantra')
    elif i == 'FCS':
        model.append('Focus')
    elif i == 'MTG':
        model.append('Mustang')
    elif i == 'ODY':
        model.append('Odyssey')
    elif i == 'PTC':
        model.append('PT Cruiser')
    elif i == 'SLV':
        model.append('Silvarado')
    else:
        model.append(np.nan)

        
df['Model (Full Name)'] = model
        
df

# Hurraaaaaaaaaaayy!!
```

```python
# The next 2 (06) in the Car ID is the production year of the car, Fill it in the Manufacture Year column

df['Manufacture Year'] = df['Car ID'].str[2:4]
df
```

```python
# Trying to convert the year column to numeric dataype

df['Manufacture Year'] = [pd.to_numeric[i] + 2000 for i in df['Manufacture Year'] if i < 22]
df

# There is an error where a string is present somewhere in the data

```

```python
# Let's trace it out (the string among the year data) and get rid of it

df['Manufacture Year'].replace('O', '', inplace = True, regex= True)

# Done!
```

```python
#Converting the year column to numerical datatype

df['Manufacture Year'] = pd.to_numeric(df['Manufacture Year']) 


'''
Because the year is in 2 digits, and some are from the 20th century, 
we will add 2000 to the 21th century values and 1900 to the 20th century values
'''

# This time around, we adopted the list comprehension method

df['Manufacture Year'] = [i + 2000 if i < 22 else i + 1900 for i in df['Manufacture Year']]
df
```

```python
# Having done that, we now calculate the age column

df.Age = 2022 - df['Manufacture Year']  # I actually did this project in 2022. Lolz, though the data was collected in 2014
df
```

```python
# Using the already provided miles covered, calculate the Miles Per Year column

df['Miles / Year'] = round(df.Miles/df.Age, 2)
df
```

```python
# The Covered column should tell whether the car has covered its warantee miles provided on the Warantee Miles column tellig if the warantee is covered or not


# To do this, I leveraged the numpy 'where' function

df['Covered?'] = np.where(df['Miles'] > df['Warantee Miles'], 'Yes', 'Not Yet')
df
```

```python
# new id should look like HY06ELA052BLU

df['New Car ID'] = df.Make + \
df['Manufacture Year'].astype(str).str[2:] + \
df['Model'] + \
df['Color'].str[:3].str.upper() + \
df['Car ID'].str[7:]


df
```
